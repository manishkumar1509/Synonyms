Get list of synonyms from response

```python
def getSynonyms(r):
  synonyms=[];
  if(r.status_code != 200):
    return [];

  for i in ['verb','adjective','noun','adverb']:
    if(i in r.json() and ('syn' in r.json()[i] or 'sim' in r.json()[i])):
      if('syn' in r.json()[i]):
        synonyms.extend(r.json()[i]['syn'])
      if('sim' in r.json()[i]):
        synonyms.extend(r.json()[i]['sim'])

  return synonyms
```

Finding the groups using a naive approach of finding common synonyms.

```python
result=[]

def add_result(i,j):
  flag=0
  for x in result:
    if(i in x):
      x.add(j)
      flag=1
  if flag==0:
    result.append({i,j})


for i in synonyms:
  i_=list(synonyms.keys()).index(i)
  for jn in range(i_+1,len(synonyms)):
    j=list(synonyms.keys())[jn]
    a_set = set(synonyms[i])
    b_set = set(synonyms[j])
    if (a_set & b_set):
        add_result(i,j)
```
